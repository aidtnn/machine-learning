```python
# import tensorflow as tf 
import tensorflow as tf
import numpy as np
config = tf.ConfigProto()
config.gpu_options.allow_growth=True
v3 = tf.placeholder(tf.float32, name="v3") 
v4 = tf.Variable(223. , name="v4") 
v5 = tf.Variable(234. , name="v5") 
d=tf.constant(22,name='d',dtype='float32')
a = tf.add(v3, d)
v2_saver = tf.train.Saver({"vg": v4})

with tf.Session(config=config) as sess:
    for step in range(2):
        sess.run(tf.global_variables_initializer())
        #b=sess.run(v2)
        tt=sess.run(a,feed_dict={v3: step})
        v6=tf.Variable(tt, name="v6")
        v2_saver.save(sess, './tung',global_step=step)
   
```
